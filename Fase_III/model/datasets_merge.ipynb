```python
# Split the df_final date column into two columns: date and time

df_final = pd.read_csv('datasets/datasetsfinal/FINALDATASET_v2.csv')

df_final['date'] = pd.to_datetime(df_final['date'])

# Create a new 'date_only' column
df_final['date_only'] = df_final['date'].dt.date

# Create a new 'time' column
df_final['time'] = df_final['date'].dt.time

# You can drop the original 'date' column if you want
df_final = df_final.drop(columns='date')

df_final.head()

df_final.to_csv('datasets/datasetsfinal/FINALDATASET_v2.csv', index=False)
```

```python
# Merge the datasets meteo.csv and FINALDATASET_v2.csv based on date and latitude and longitude

df_meteo = pd.read_csv('datasets/datasetsfinal/meteo.csv')

df_total = pd.read_csv('datasets/datasetsfinal/FINALDATASET_v2.csv')

dates1 = set(df_meteo['date'])
dates2 = set(df_total['date_only'])

# Find the common dates
common_dates = dates1.intersection(dates2)

print(len(common_dates))
```

```python
# Filter the datasets by the common dates

df_meteo = df_meteo[df_meteo['date'].isin(common_dates)]

df_total = df_total[df_total['date_only'].isin(common_dates)]
```

```python
# Rename the column date_only to date to be able to merge the datasets

df_total.columns = ['longitude','latitude','no2','pm10','pm25','pm1','date','time']

# Merge the datasets on date but keep all hours of the day (if they do not exist put null values)

df_final = pd.merge(df_total, df_meteo, on=['date', 'latitude', 'longitude'], how='outer')

df_final.head()

df_final.to_csv('datasets/datasetsfinal/FINALDATASETmerged_v2.csv', index=False)
```

```python
# For the dataset final, if the day is the same the atributes tempmax,tempmin,temp,feelslike,dew,humidity,severerisk,precip,snow,windspeed,sealevelpressure,cloudcover,visibility,solarradiation,solarenergy,uvindex are the same

df = pd.read_csv('datasets/datasetsfinal/FINALDATASETmerged_v2.csv')

# Combine 'date' and 'time' into a single datetime column to ensure the filling happens in chronological order
df['datetime'] = pd.to_datetime(df['date'] + ' ' + df['time'])
df = df.sort_values('datetime')

# List of columns to fill
columns_to_fill = ['tempmax', 'tempmin', 'temp', 'feelslike', 'dew', 'humidity', 'severerisk', 
                   'precip', 'snow', 'windspeed', 'sealevelpressure', 'cloudcover', 'visibility', 
                   'solarradiation', 'solarenergy', 'uvindex']

# Apply forward fill
for col in columns_to_fill:
    df[col] = df[col].fillna(method='ffill')

# You can also backward fill for any remaining nulls if needed
for col in columns_to_fill:
    df[col] = df[col].fillna(method='bfill')

# Separate the 'datetime' column back into 'date' and 'time' columns

df['date'] = df['datetime'].dt.date

df['time'] = df['datetime'].dt.time

# Drop the 'datetime' column

df = df.drop(columns='datetime')

# Transform the date column to datetime

df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d')

# Do the same with the time column

df['time'] = pd.to_datetime(df['time'], format='%H:%M:%S')

df.head()

df.to_csv('datasets/FINALDATASET.csv', index=False)
```
