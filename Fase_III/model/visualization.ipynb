## Exploração de Dados

### Duplicados e Valores nulos

```python
print(str(braga_data[braga_data.duplicated()])) # Verifica se e quais são os valores duplicados
print("\n")
print(str(braga_data.isna().any())) # Verifica se há valores nulos no dataset
```

```python
print(braga_data.isna().sum()) # Contagem dos valores em falta do dataset
```

```python
sns.heatmap(braga_data.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
# Identificar o Intervalo de Latidude e Longitude do dataset
print("Latitude:\n", braga_data['latitude'].min(), braga_data['latitude'].max())
print("Longitude:\n", braga_data['longitude'].min(), braga_data['longitude'].max())

# dar plot às coordenadas num mapa para ver se estão corretas
plt.figure(figsize=(5,5))
sns.scatterplot(x='longitude', y='latitude', data=braga_data)

```

```python
# Identificar o Intervalo de Tempo
# Converter a coluna 'data' em datetime
braga_data['date'] = pd.to_datetime(braga_data['date'],format="%Y-%m-%d")
print("Time:\n", braga_data['date'].min(), braga_data['date'].max())
```

```python
# Gráfico de linhas para ver a evolução dos poluentes ao longo do tempo
# Há uma coluna para cada poluente
pollutants = ['no2', 'pm10', 'pm25', 'pm1']
plt.figure(figsize=(20,10))
for i in range(len(pollutants)):
    plt.subplot(2,2,i+1)
    sns.lineplot(x='date', y=pollutants[i], data=braga_data)
    plt.title(pollutants[i])
    plt.xticks(rotation=45)
plt.tight_layout()
```

```python
# obter todas as colunas menos a dos poluentes
columns = braga_data.columns
columns = columns.drop(pollutants)
# remover a coluna 'date' e 'time'
columns = columns.drop(['date', 'time', 'latitude', 'longitude'])
print(columns)

plt.figure(figsize=(20,10))
# 2 colunas e 3 linhas
for i in range(len(columns)):
    plt.subplot(9,2,i+1)
    sns.lineplot(x='date', y=columns[i], data=braga_data)
    plt.title(columns[i])
    plt.xticks(rotation=45)
plt.tight_layout()
```

## Outliers

```python
sns.set_theme(palette="pastel")
# plot outliers boxplot para cada coluna
for column in braga_data.columns:
    if column not in ['date', 'time', 'latitude', 'longitude']:
        plt.figure(figsize=(5,5))
        sns.boxplot(x=braga_data[column])
```

```python
# Theoretical Quantile-Quantile Plots
def plot_probplots(columns, df):
    for column in columns:

        plt.figure()
        stats.probplot(df[column], dist="norm", plot=plt)
        plt.title(f"Probability plot of {column}")
    plt.show()

columns = braga_data.columns.drop(['date', 'time', 'latitude', 'longitude'])
plot_probplots(columns,braga_data)
```

## Correlações

```python
mask = np.triu(np.ones_like(braga_data.corr(), dtype=bool))
sns.heatmap(braga_data.corr(), mask = mask, annot = True, cmap='coolwarm')
# diminuir o tamanho da letra da correlação
plt.rcParams['font.size'] = 40
# só quero a superior à diagonal
mask = np.triu(np.ones_like(braga_data.corr(), dtype=bool))
plt.tick_params(axis='both', labelsize=20, labelbottom = True, bottom=False, top = False, labeltop=False)
# reduzir simetria
plt.xticks(rotation=45)
plt.yticks(rotation=0)
# aumentar o tamanho da figura
plt.rcParams['figure.figsize'] = [40, 20]
plt.show()
```
